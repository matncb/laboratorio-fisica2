```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
m_calorimetro = 1101.52 #massa calorimetro
```

```python
def grafico(T_data, t2 = False):
    t = np.arange(0,380 + 20, 20)
    if t2: t = np.arange(0,220 + 20, 20)
    plt.errorbar(t, T_data,yerr=0.1, ls= '', marker = 'o', ms = 3)
    plt.xlabel("t(s)")
    plt.ylabel("T(°C)")
    plt.title("Temperatura x Tempo")
```

```python
#Exp I

m_c_a = 1200.45 #massa calorimetro + agua
m_a = m_c_a - m_calorimetro #massa agua

m_base = 1.808 #massa NaOH
mols_base = m_base/39.997 #mols NaOH

m_s = m_a + m_base #massa da solução
c_s = 3.97

T_i = 23.9
T_f = 27.5
delta_H = -m_s*c_s*(T_f -T_i)
delta_H_mol = delta_H/mols_base

print(f"Delta H = {delta_H} J")
print(f"Delta H = {delta_H_mol} J/mol")
```

```python
T_data = np.array([23.9, 24.5, 24.9, 25.3, 25.8, 26.1, 26.4, 26.8, 27.0, 27.2, 27.3, 27.4, 27.4,27.5, 27.5, 27.5, 27.5, 27.4, 27.4, 27.4])
grafico(T_data)
```

```python
#Exp II

m_c_acido = 1201.52 #massa calorimetro + HCl
m_acido = m_c_acido - m_calorimetro #massa  solucao HCl
mols_acido = m_acido/(1020)

m_base = 1.824 #massa NaOH
mols_base = m_base/39.997 #mols NaOH

mols_reacao = min([mols_acido, mols_base])

m_s = m_a + m_base #massa da solução
c_s = 3.97

T_i = 24.9
T_f = 33.4
delta_H = -m_s*c_s*(T_f -T_i)
delta_H_mol = delta_H/mols_reacao

print(f"Delta H = {delta_H} J")
print(f"Delta H = {delta_H_mol} J/mol")
```

```python
T_data = np.array([24.9, 26.8, 28.7, 30.0, 31.3, 32.3, 32.8, 33.3, 33.3, 33.4, 33.4, 33.3, 33.3,33.2, 33.1, 33.0, 33.0, 33.0, 33.0, 32.9])
grafico(T_data)
```
