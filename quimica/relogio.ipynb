```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])

from scipy.optimize import  curve_fit
from sklearn.metrics import r2_score
```

```python
V1 = np.array([5, 10, 15, 20, 25])
t = np.array([60 + 25.54, 43.46, 28.26, 21.60, 17.91])

x_data = V1
y_data = 1/t
```

```python
f = lambda x, a, b : a*x +b*0

popt, pcov = curve_fit(f, x_data, y_data)
a = popt[0]
b = popt[1]
a_err = pcov[0][0]
r2 = r2_score(y_data, f(x_data, a, b))

a
```

```python
plt.scatter(x_data, y_data, label = "Pontos Experimentais")

x = np.linspace(5,25, 100)
y = f(x, a, b)

plt.plot(x,y, color = "green", label = f"Fit: $a$ = {a:.5f} ; $ r^2$ =  {r2:.3f}")

plt.ylabel("$1/{\Delta t} \ (s^{-1})$")
plt.xlabel("$V_1 \ (mL)$")
plt.title("$1/{\Delta t}$ x $V_1$")
plt.legend()
```
