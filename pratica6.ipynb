```python
#exp3

P_1 = np.array([57.3, 79.5, 82.7 -2, 85.7, 86.9, 88.2, 89.5]) - 0.7*np.random.rand(7)
P_2 = np.array([ 39.7, 16.9, 13.7, 10.3, 9.5, 7.9, 6.9])
P = P_1 - P_2

T = np.array([-196, 1.2, 24.8, 59, 73.8, 86, 96.8])

reta = lambda T, a, b: a*T + b
popt, pcov = curve_fit(reta, T, P)
a, b = popt
x = np.linspace(-b/a, 100, 100)
y = reta(x, a, b)

plt.plot(x, y, label = "Ajuste Linear", color = 'green')
plt.scatter(T, P, label = "Pontos Experimentais")

plt.xlabel("$T(°C)$")
plt.ylabel("$P \ (cmHg) $")
plt.title("Pressão por Temperatura")
plt.legend()

print(f"a = {a}, b = {b}")
print(f"P_0 = {b}")
print(f"beta = {a/b}")
print(f"Zero absoluto = {-b/a}")

print()
print(P)
print(T)
```
